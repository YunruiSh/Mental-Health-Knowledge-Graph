```python
import pandas as pd
raw = pd.read_csv("raw_wiki_Q12135.csv")
raw.drop_duplicates(inplace = True)
raw.head()
```

```python
raw_drugs = pd.read_csv("raw_wiki_drugs.csv")
raw_drugs.drop_duplicates(inplace = True)
raw_drugs.head()
```

```python
raw_symps = pd.read_csv("raw_symptoms.csv")
raw_symps.drop_duplicates(inplace = True)
raw_symps.head()
```

```python
# all symptoms
allSymptoms = {'symptoms': raw_symps["symptoms"], 'symptomsLabel':raw_symps["symptomsLabel"]}
allSymptoms_csv = pd.DataFrame(data=allSymptoms)
allSymptoms_csv.drop_duplicates(inplace = True)
allSymptoms_csv.to_csv("allSymptoms_csv.csv", sep=',', encoding='utf-8')
```

```python
# all diseases
alldiseases = {'item': raw["item"], 'itemLabel':raw["itemLabel"]}
alldiseases_csv = pd.DataFrame(data=alldiseases)
alldiseases_csv.drop_duplicates(inplace = True)
alldiseases_csv.to_csv("alldiseases_csv.csv", sep=',', encoding='utf-8')
```

```python
raw_riskFs = pd.read_csv("raw_riskFactors.csv")
raw_riskFs.drop_duplicates(inplace = True)
raw_riskFs.head()
```

```python
# try
loc = raw_drugs.loc[raw_drugs["item"] == "http://www.wikidata.org/entity/Q161790"]
for i in loc["drugsLabel"]:
    print(i)
```

```python
itemList = []
for i in raw_drugs["item"]:
    if i not in itemList:
        itemList.append(i)
```

```python
drugList = []
for i in itemList:
    locI = raw_drugs.loc[raw_drugs["item"] == i]
    save_dl = []
    for j in locI["drugsLabel"]:
        save_dl.append(j)
    drugList.append(save_dl)
```

```python
drugList2 = []
for i in itemList:
    locI = raw_drugs.loc[raw_drugs["item"] == i]
    save_dl = []
    for j in locI["drugs"]:
        save_dl.append(j)
    drugList2.append(save_dl)
```

```python
newLabel = {'item': itemList, 'drugs':drugList2, 'drugsLabel':drugList}
sort_wiki_drugs = pd.DataFrame(data=newLabel)
sort_wiki_drugs.head()
```

```python
print(len(sort_wiki_drugs))
```

```python
itemList = []
for i in raw_symps["item"]:
    if i not in itemList:
        itemList.append(i)
        
sympsList = []
for i in itemList:
    locI = raw_symps.loc[raw_symps["item"] == i]
    save_dl = []
    for j in locI["symptomsLabel"]:
        save_dl.append(j)
    sympsList.append(save_dl)
    
sympsList2 = []
for i in itemList:
    locI = raw_symps.loc[raw_symps["item"] == i]
    save_dl = []
    for j in locI["symptoms"]:
        save_dl.append(j)
    sympsList2.append(save_dl)
    
newLabel = {'item': itemList, 'symptoms':sympsList2, 'symptomsLabel':sympsList}
sort_wiki_symps = pd.DataFrame(data=newLabel)
sort_wiki_symps.head()
```

```python
print(len(sort_wiki_symps))
```

```python
itemList = []
for i in raw_riskFs["item"]:
    if i not in itemList:
        itemList.append(i)
        
riskFsList = []
for i in itemList:
    locI = raw_riskFs.loc[raw_riskFs["item"] == i]
    save_dl = []
    for j in locI["riskFactorsLabel"]:
        save_dl.append(j)
    riskFsList.append(save_dl)
    
riskFsList2 = []
for i in itemList:
    locI = raw_riskFs.loc[raw_riskFs["item"] == i]
    save_dl = []
    for j in locI["riskFactors"]:
        save_dl.append(j)
    riskFsList2.append(save_dl)
    
newLabel = {'item': itemList, 'riskFactors':riskFsList2, 'riskFactorsLabel':riskFsList}
sort_wiki_riskFs = pd.DataFrame(data=newLabel)
sort_wiki_riskFs.head()
```

```python
print(len(sort_wiki_riskFs))
```

```python
sort_wiki = pd.merge(raw, sort_wiki_drugs, how="left")
sort_wiki = pd.merge(sort_wiki, sort_wiki_symps, how="left")
sort_wiki = pd.merge(sort_wiki, sort_wiki_riskFs, how="left")
sort_wiki.head()
```

```python
sort_wiki.to_csv("sort_wiki_Q12135.csv", sep=',', encoding='utf-8')
```
