Deal with Person-related Tables

```python
import pandas as pd
# read csv files
bf_therapist = pd.read_csv("bf_therapist.csv", sep="\t", on_bad_lines="skip")
# drop first column
first_col = bf_therapist.columns[0]
bf_therapist = bf_therapist.drop([first_col], axis=1)
# deal with mid csv
bf_therapist = pd.read_csv("mid.csv", sep=",", on_bad_lines="skip")
# drop first column
first_col = bf_therapist.columns[0]
bf_therapist = bf_therapist.drop([first_col], axis=1)
bf_therapist.head()
```

```python
# print(bf_therapist["street"][1])
```

```python
# merge street, city, postalcode => address
address = []
for line in range(len(bf_therapist)):
    street = ""
    city = ""
    postcode = ""
    if bf_therapist["city"][line] is None:
        street = ""
    else: 
        street = str(bf_therapist["street"][line]) + ", "
    if bf_therapist["city"][line] != None:
        street = street + str(bf_therapist["city"][line]) + ", "
    if bf_therapist["postalcode"][line] != None:
        street = street + str(bf_therapist["postalcode"][line])
    address.append(street+city+postcode)
new_address = []
for line in address:
    tmp = ""
    if line[:5] == "nan, ":
        tmp = line[5:]
    else:
        tmp = line
    new_address.append(tmp)
address = new_address        
```

```python
# print(bf_therapist["name"])
for n in bf_therapist:
    print(n)
```

```python
person_id = []
for i in range(9956):
    person_id.append(i)
```

```python
# new df
new_label = {"person_id": person_id,
             "name": bf_therapist["name"], 
             "title": bf_therapist["title"], 
             "about": bf_therapist["about"], 
             "website": bf_therapist["website"], 
             "mobile": bf_therapist["mobile"], 
             "address": address, 
             "geo_location": bf_therapist["geo"], 
             "ethnicity": bf_therapist["ethnicity"], 
             
             "therapy_type": bf_therapist["therapy_type"], 
             "age": bf_therapist["age"], 
             "communities": bf_therapist["communities"], 
             "modality": bf_therapist["modality"], 
            
             "specialities": bf_therapist["specialities"], 
             "specialties_detail": bf_therapist["specialties_detail"]}
therapist = pd.DataFrame(data=new_label)
therapist.head()
```

```python
# therapy_type
t_type = pd.read_csv("therapy_type.txt", sep="\n", header=None, on_bad_lines="skip")
t = []
for n in t_type[0]:
    t.append(n)
t_type = t
print(len(t))

#
max = 0
for line in therapist["therapy_type"]:
    count = 0
    for i in t_type:
        if i in str(line):
            count += 1
    if count > max:
        max = count
print(max)
```

```python
# find person_id list for therapy_type
therapy_id = []
for i in range(51):
    therapy_id.append(i)
therapy_label = {"therapy_id": therapy_id, 
                 "therapy_type": t_type}
therapy_type_df = pd.DataFrame(data=therapy_label)
therapy_type_df.head()
```
